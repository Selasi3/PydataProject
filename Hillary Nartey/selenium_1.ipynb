```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
driver
```

```python
url = "https://www.jumia.com.gh/laptops"
```

```python
#launches  a browser
driver.implicitly_wait(10)
driver.get(url)
```

```python
#find product name by Xpath
product = [name.text for name in driver.find_elements_by_xpath('//h3')]
product
```

```python
price = [price.text for price in driver.find_elements_by_xpath('//div[@class="info"]//child::div[@class="prc"]')]
price
```

```python
#find links by xpath
link = [url.get_attribute("href") for url in driver.find_elements_by_xpath("//article[contains(@class,'prd _fb')]//child::a")]
```

```python
link
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "product" : product,
    "price" : price,
    "link" : link
    })
df
```

```python
df.to_csv('selenium_1.csv')
```
