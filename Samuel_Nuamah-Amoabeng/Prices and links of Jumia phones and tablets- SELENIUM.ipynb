```python
from selenium import webdriver
```

```python
jurl = 'https://www.jumia.com.gh/phones-tablets/'
drivers = webdriver.Chrome()
drivers.get(jurl)
```

```python
jptabs = [ptabs.text for ptabs in drivers.find_elements_by_class_name('name')]
jptabs
```

```python
jptabs1 = [ptabs.text for ptabs in drivers.find_elements_by_xpath("//div[@class= 'name']")]
jptabs1
```

```python
len(jptabs1)
```

```python
jptabs_prices = [ptabs.text for ptabs in drivers.find_elements_by_class_name('prc')]
jptabs_prices
```

```python
jptabs_prices1 = [price.text for price in drivers.find_elements_by_xpath("//div[@class= 'prc']")]
jptabs_prices1
```

```python
len(jptabs_prices1)
```

```python
jptabs_link1 = []
jptabs_link = drivers.find_elements_by_xpath("//*[@id='jm']/main/div[2]/div[3]/section/div[1]/article/a")


for i in jptabs_link:
    jptabs_links = print(i.get_attribute('href'))
    jptabs_link1.append(i.get_attribute('href'))
print(jptabs_link1)
```
