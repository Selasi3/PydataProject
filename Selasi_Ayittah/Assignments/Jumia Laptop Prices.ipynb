```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup as bs
import bs4
import pandas as pd
```

```python
url = 'https://www.jumia.com.gh/laptops/'
request = Request(url, headers= {"User-Agent":"Mozilla/5.0"})
```

```python
webpage = urlopen(request).read()
```

```python
result =bs(webpage)
```

```python
laptops =result.select(".info .name")
laptops
```

```python
laptopprice=result.select(".info .prc")
```

```python
laptopname=[txt.text.strip() for txt in laptops]
```

```python
Price = [txt.text.strip() for txt in laptopprice]
```

```python
df= pd.DataFrame({"Name":laptopname,"Price":Price})
```

```python
df
```

```python
solution = df[:40]
```
