```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
```

```python
url = "https://www.jumia.com.gh/phones-tablets/"
headers = {
        "User-Agent": "Mozilla/6.0"
}
```

```python
req = Request(url, headers=headers)
web_page = urlopen(req).read()
# print(web_page)
```

```python
page = soup(web_page, 'html.parser')
print(page.prettify())
```

```python
cards_container = page.find('div', class_="-paxs row _no-g _4cl-3cm-shs") 
# print(cards_container.prettify())
# print(type(cards_container))
```

```python
product_names_with_tags = cards_container.find_all("h3", class_="name")
# print(product_names)
product_names = [names.text.strip() for names in product_names_with_tags if names.text.strip() != ""]
# print(len(product_names))
```

```python
product_prices_with_tags = cards_container.find_all("div", class_="prc")
# print(product_prices_with_tags)
product_prices = [prices.text.strip() for prices in product_prices_with_tags if prices.text.strip() != ""]
print(len(product_prices))
```

```python
product_names_prices = [(product_names[i], product_prices[i]) for i in range(len(product_names))]
print(product_names_prices)
```

```python
import pandas as pd
columns = ["Product", "Price"]
df = pd.DataFrame(product_names_prices, columns=columns)
df.head()
```

```python
df.to_csv("Jumia-Tablets-and-Phones-catelog.csv", index=False)
```
