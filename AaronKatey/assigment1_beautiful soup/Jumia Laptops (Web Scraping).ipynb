```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup as sp

```

```python
url = 'https://www.jumia.com.gh/laptops/'
request = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
page = urlopen(request).read()

page
```

```python
web_page_ = sp(page, "lxml")
products = [pro.get_text() for pro in web_page_ .select(".info .name")]
price = [cost.get_text() for cost in web_page_ .select(".info .prc")]

import pandas as pd

df = pd.DataFrame({
    "Laptop": products[:40],
    "Price": price[:40]
    
})
df 
```

```python
df.to_csv("Laptops and their Prices on Jumia.csv")
```
