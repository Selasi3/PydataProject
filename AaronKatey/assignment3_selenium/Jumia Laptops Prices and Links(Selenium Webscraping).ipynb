```python
url = 'https://www.jumia.com.gh/laptops'
driver = webdriver.Chrome()

driver.get(url)
```

```python
laptops = [names.text for names in driver.find_elements_by_xpath('//h3')]
laptops
```

```python
lap_cost= [prices.text for prices in driver.find_elements_by_xpath('//div[@class= "info"]//child::div[@class="prc"]')]
lap_cost
```

```python
lap_links= [links.get_attribute("href") for links in driver.find_elements_by_xpath('//article[contains(@class,"prd _fb")]//child::a')]
lap_links
```

```python
import pandas as pd

all_laptops = pd.DataFrame({
    "Laptops": laptops,
    "Prices": lap_cost,
    "Links":lap_links
    
})
all_laptops
```

```python
all_laptops.to_csv('Laptop Prices and Links on Jumia.csv')
```
